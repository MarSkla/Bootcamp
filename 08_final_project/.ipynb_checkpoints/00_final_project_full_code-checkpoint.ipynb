```python
# sprawdzilości obserwacji odstających, metodą 1.5 IQR
# z danych usuwam 'target' - tutaj nie ma outliers
df_out = df.drop('target', axis = 1)

quar_1 = df_out.quantile(0.25)
quar_3 = df_out.quantile(0.75)
iqr = quar_3 - quar_1
```

```python
left_border = (quar_1 - 1.5 * iqr)
right_border = (quar_3 + 1.5 * iqr)
outliers_left_no = (df_out[iqr.index] < left_border).sum()
outliers_right_no = (df_out[iqr.index] > right_border).sum()
outliers = pd.DataFrame({'left_border':left_border, 'right_border':right_border,
                         'outliers_left_no':outliers_left_no,
                         'left_all_prc': round(outliers_left_no/df.shape[0],3),
                         'outliers_right_no':outliers_right_no,
                         'right_all_prc': round(outliers_right_no/df.shape[0],3)})
outliers
```

<br> hipoteza zerowa: nie ma zależności pomiędzy poszczególnymi zmiennymi kategorycznymi
<br> hipoteza alternatywna: są zależności pomiędzy poszczególnymi zmiennymi kategorycznymi

```python
# w pierwszej kolejności, zmieniam czasowo typ zmiennej 'target' z numerycznego, na jakościowy
df_chi = df.copy()

df_chi['target'] = df_chi['target'].replace({1: 'repaid', 0: 'lost'})
df_chi['target'] = df_chi['target'].astype('category')
```

```python
# tworzę klasę do parametrów chi-square

class chi_square:

    def __init__(self, frame, sig_lev):
        self.frame = frame
        self.sig_lev = sig_lev
        
        d_f = self.frame
        chosen_col = d_f.select_dtypes('category')
        self.columns = type_cat_col = chosen_col.columns.tolist()
        
        #każdy parametr zwraca nowy data frame pokazujący zależności pomiedzy poszczególnymi kolumnami
        self.chi2 = chi2_df = pd.DataFrame(columns = [type_cat_col],
                                           index = [type_cat_col]) 
        
        self.p_value = p_df = pd.DataFrame(columns = [type_cat_col],
                                           index = [type_cat_col])
        
        self.dof = dof_df = pd.DataFrame(columns = [type_cat_col],
                                         index = [type_cat_col])
        
        self.crival = chi_critival_value = pd.DataFrame(columns = [type_cat_col],
                                                        index = [type_cat_col])
        
        self.chi_result = pd_result = pd.DataFrame(columns = [type_cat_col],
                                               index = [type_cat_col])
        
        self.cramer = pd_cramer = pd.DataFrame(columns = [type_cat_col],
                                               index = [type_cat_col])
        
        self.final_result = pd_final_result = pd.DataFrame(columns = [type_cat_col],
                                               index = [type_cat_col])

    
        for x in type_cat_col:
            data_b = d_f[x]
            for y in type_cat_col:
                if y == x:
                    continue
                else:
                    data_a = d_f[y]
                    # dla automatyzacji korzystam z chi2_contingency, więć potrzebuję tymczasowe tabele
                    temp_table = pd.crosstab(data_a,data_b)
                    
                    #obliczam poszcególne paramtery chi-square
                    chi2, p, dof, expected = chi2_contingency(temp_table)
                    
                    # umieszcam wyniki do poszczególnych 'data frames'
                    p_df.loc[y, x] = round(p,4)
                    
                    dof = dof_df.loc[y, x] = float(round(dof,2))
                    
                    chi = chi2_df.loc[y, x] = float(round(chi2,2))
                    
                    crit_val = chi_critival_value.loc[y, x] = float(scipy.stats.chi2.ppf(sig_lev, dof))
                    
                    pd_res = pd_result.loc[y, x] = ([str('yes') if crit_val < chi else 'no'])     
                    
                    #pd_cramer.loc[y, x] = pd_result.loc[y, x]
                    
                    v_cramer_denominator = (d_f.shape[0]*d_f.shape[1]) * (min(temp_table.shape[0] - 1, temp_table.shape[1]- 1))
                    cram = pd_cramer.loc[y, x] = round(math.sqrt(chi/v_cramer_denominator),2)
                    
                    pd_final = pd_final_result.loc[y, x] = ([str('yes') if p < cram else 'no'])
                    
                    
                    
                    
        #([(math.sqrt(chi/(pd_result.shape[0] * (min(pd_result.shape[0], pd_result.shape[1] -1))))) for pd_result.loc[y, x] == str('yes')])
                    
#                    pd_cramer.loc[y, x] = [(math.sqrt(chi/(pd_result.shape[0] * (min(pd_result.shape[0], pd_result.shape[1] -1))))) if pd_result.loc[y, x] == 'yes']

        print(f'\tvariables (all "category" type) chosen to conduct chi2 test:')
        print(*self.columns, sep = '\n')

        
    def chi_correlation(self): 
        
#        print(f'dependency between variables:\n')
        return self.chi_result


    def chi2_chi2(self):
        
#        print(f'\n\tcrosstab - "chi2" results (% of occurence):\n')
        return self.chi2
    
    def chi2_p(self):

#        print(f'\n\tcrosstab - "probability":\n')
        return self.p_value
    
    def chi2_dof(self):

#        print(f'\n\tcrosstab - "degrees of freedom":\n')
        return self.dof

    def chi2_crtical_value(self):

#        print(f'\n\tcrosstab - "critical values - significance level {self.sig_lev}":\n')
        return self.crival
    
    def significance(self):
        

        return self.cramer
    
    def result(self):
        
        return self.final_result
```

```python
# funkcj|a zmieniająca wartośc fico na przedział zgodnie z opisem 'FICO+score+ranges.odf'
def fico_grade(x):
    if x < 580:
        return '1_Poor'
    elif x >= 580 and x <= 699:
        return '2_Fair'
    elif x >= 670 and x <= 739:
        return '3_Good'
    elif x >= 740 and x <+ 799:
        return '4_Very good'
    elif x >= 800:
        return '5_Exceptional'
```

```python
#tworzę tymczasowy df tylko z obserwacjami gdzie był default do zmiany fico na przedział do którego należy
df_temp_q2_2_def = df[df['target'] == 0][['fico', 'credit_hist', 'target']]
# nowa kolumna z info o przedziale fico
df_temp_q2_2_def['defaults_no'] = df_temp_q2_2_def['fico'].apply(fico_grade)

#nowy df do kalkulacji wystąpień defaultu w poszczególnych gdupach fico
df_work_q2_def = pd.DataFrame(df_temp_q2_2_def['defaults_no'].value_counts()).reset_index()
df_work_q2_def
```

```python
#tworzę tymczasowy df tylko z obserwacjami gdzie była spłata do porównania z powyższą
df_temp_q2_2_rep = df[df['target'] == 1][['fico', 'credit_hist', 'target']]
# nowa kolumna z info o przedziale fico
df_temp_q2_2_rep['repayment_no'] = df_temp_q2_2_rep['fico'].apply(fico_grade)

#nowy df do kalkulacji wystąpień defaultu w poszczególnych gdupach fico
df_work_q2_rep = pd.DataFrame(df_temp_q2_2_rep['repayment_no'].value_counts()).reset_index()
df_work_q2_rep
```

```python
# łącze obydwie, żeby policzyc procent niespłącjących w kazdej grupie 'fico'
df_final_q2 = pd.merge(df_work_q2_def, df_work_q2_rep, how = 'left', on = 'index')
df_final_q2['default_prc'] = df_final_q2['defaults_no']/(df_final_q2['repayment_no'] + df_final_q2['defaults_no'])
# sortuję wg kolejności fico (rosnąco)
df_final_q2 = df_final_q2.sort_values('index')
df_final_q2
```

```python
# tworzę nowy data frame 'new_features'
new_features = pd.DataFrame()
```

```python
# pierwszą zmienną jest wcześniej stworzona 'credit_hist'
# wskazuje ilość miesięcy pomiedzy 'earliest_cr_line' a 'issue_d'
new_features['credit_hist'] = new_features_wip['credit_hist']
```

```python
# drugą zmienną będzie informacja, czy kwota otrzymana 'loan'
# jest równa wnioskowanej'loan_amnt'
# trzecią zaś, otrzymany % wnioskowanej kwoty

new_features['granted_requested'] = np.where(new_features_wip['loan'] == new_features_wip['loan_amnt'], 'yes', 'no')
new_features['granted_requested'].astype('category')
new_features['granted_requested_prc'] = round(new_features_wip['loan']/new_features_wip['loan_amnt'],2).astype('float')
```

```python
# czwartą nowo tworzoną cechą, będzie miesiąc przyznania pożyczki
# piąta cechą będzie rok przyznanani pozyczki

new_features['month_granted'] = new_features_wip['issue_d'].dt.month.astype('int')
new_features['year_granted'] = new_features_wip['issue_d'].dt.year.astype('int')
```

```python
# szóstą będzie cyfra z subgrade (powtarza się dla kazdej litery)

new_features['grade_subset'] = new_features_wip['sub_grade'].str[1].astype('int')
```

```python
# siódmą, info ile procent pozyczki sfinansowali inwestorzy

new_features['inv_share_loan_prc'] = round(new_features_wip['funded_amnt_inv']/new_features_wip['loan'], 5).astype('float')
```

```python
# ósma to % wskazujący wysokośc otrzymanej pożyczki vs roczny dochód
new_features['loan_icome_prc'] = round(new_features_wip['loan']/new_features_wip['annual_inc'], 5).astype('float')
```

```python
# dziewiąta to przypisanie do obserwacji średniej dla stanu zamieszkania
# pozyczkobiorcy
# dziesiąta to porównanie wartości pożyczki ze średnią w danym stanie

# najpierw liczę średnia dla kazdego stanu
state_average = new_features_wip[['state', 'loan']].groupby('state').mean().reset_index()
# i wynik zamieniam w słownik do mapowania
state_average_dict = dict(zip(state_average['state'],state_average['loan']))

# tworzę dziewiątą zmienną przypisując wartości ze słwonika
new_features['state_avg'] = new_features_wip['state'].map(state_average_dict).astype('float').astype('float')

# do porównania dodaje równiez zmienną loan, którą nastepnię usunę aby nie duplikować z 'df' 
new_features['loan'] = new_features_wip['loan'].astype('float')

# przypisuję wynik porównania
new_features['state_avg_desc'] = np.where(new_features['state_avg'] < new_features['loan'], 'higher', 'lower')
new_features['state_avg_desc'].astype('category')

# usuwam tymczasowa zmienną 'loan'
new_features = new_features.drop('loan', axis = 1)
```

```python
# jedeasta zmienna całkowity dostępny limit kredytowy skalkulowany na podstawie
# 'revol_bal' oraz 'revol_util'
# 'revol_util' dziele dodatkowo przez 100, pamietając że usunąłem %

new_features['total_credit_available'] = np.where((new_features_wip['revol_bal'] > 0)
                                                  & (new_features_wip['revol_util'] > 0)
                                                  ,round((new_features_wip['revol_bal']
                                                         /(new_features_wip['revol_util']
                                                           /100)), 2)
                                                  , 0)
```

```python
# dwunasta zmienna to miesięczna spłąta zadłużenia (poza pozyczką)
# na podstawie wskaźnika dti i rocznego dochodu podanego przez pożyczkobiorcę
# przyjmuję, iż skoro wzięte są raty miesięczne, wtedy dochód też jest dzielony przze 12
new_features['monthly_repayments_other_debt'] = round(((new_features_wip['dti_ratio']/100)*(new_features_wip['annual_inc']/12)).astype('float'), 2)
```

```python
# trzynasta, to info ile środków z rocznego wynagrodzenia zostaje po obsłudze dotychczasowego długu (ammualizowany)
new_features['free_cash_annualy'] = round(new_features_wip['annual_inc'] - (new_features['monthly_repayments_other_debt'] *12), 2).astype(float)
```

```python
# czternasta, to info ile gotówki zostanie rocznie po dołaczeniu rocznych rat spłaty pozyczki
new_features['free_cash_remaining'] = round(new_features['free_cash_annualy']
                                            - ((new_features['monthly_repayments_other_debt'] *12)
                                               + (new_features_wip['installment'] * 12))
                                            , 2).astype(float)
```

```python
# pietnasta zmienna, to infoamcja o ile wzrośnie łączny dług po udzieleniu wnioskowanej kwoty pozyczki
new_features['total_debt_increase_prc'] = np.where((new_features['total_credit_available'] > 0)
                                                   , round((((new_features['total_credit_available']
                                                              + new_features_wip['loan_amnt'])
                                                             /new_features['total_credit_available'])
                                                            - 1) * 100, 1).astype('float'), 0)
```

```python
# szesnasta, to info ile procent rocznego dochodu, będzie stanowił łączny dług po udzieleniu wnioskowanej kwoty pożyczki
new_features['total_debt_income_share'] = round(((new_features['total_credit_available']
                                                  + new_features_wip['loan'])
                                                 / new_features_wip['annual_inc'])* 100
                                                , 2).astype('float')
```

```python
# siedemnasta to łączny koszt pozyczki (suma rat - kwota pożyczki)
new_features['total_cost'] = round((new_features_wip['installment'] * new_features_wip['term'].astype('int')) - new_features_wip['loan'], 2)
```

```python
# w takim razie osiemnasta, to naturalnie całkowity koszt w %
new_features['total_cost_prc'] = round(((new_features_wip['installment']
                                         * new_features_wip['term']
                                         .astype('int'))
                                        / new_features_wip['loan']) - 1, 2)
```

```python
# dziewietnasta cecha info faktycznej ilości miesięcy korzystania ze spłaconej finalnie pożyczki
new_features['real_tenor'] = np.where(new_features_wip['target'] == 1,
                                      ((new_features_wip['last_pymnt_d']
                                        - new_features_wip['issue_d'])
                                       /np.timedelta64(1, 'M')), 0)
new_features['real_tenor'] = new_features['real_tenor'].astype('int')
```

```python
# dwudziesta to info, czy była wcześniejsza spłata
new_features['early_rpmnt'] = np.where((new_features_wip['target'] == 1)
                                       & (new_features['real_tenor']
                                          < (new_features_wip['term']).astype('int'))
                                       , 'yes'
                                       , 'no')
```

```python
# dwudziesta pierwsza to % kwoty pozyczki pozostajacy do spłaty w przypadku pożyczki niespłaconej 
new_features['remaining_principal_prc'] = np.where(new_features_wip['target'] == 0, round(1 - new_features_wip['total_rec_prncp']/new_features_wip['loan'], 2), 0)
```
